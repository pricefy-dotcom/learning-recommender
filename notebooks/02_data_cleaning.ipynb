```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("../data/raw/Maths.csv")
print("Data loaded")
print(f"Shape: {df.shape}")
```

```python
print("Missing values per column:")
print(df.isnull().sum())
```

```python
print(f"Duplicate rows: {df.duplicated().sum()}")
```

```python
# Convert yes/no to 1/0
binary_columns = ['schoolsup', 'famsup', 'paid', 'activities',
                  'nursery', 'higher', 'internet', 'romantic']

for col in binary_columns:
    df[col] = df[col].map({'yes': 1, 'no': 0})

print("Binary columns converted")
df[binary_columns].head()
```

```python
print("Grade Statistics:")
print(df[['G1', 'G2', 'G3']].describe())
```

```python
df.to_csv('../data/processed/Maths_cleaned.csv', index = False)
print("Cleaned data saved.")
```

```python
df = df.drop_duplicates()
print(f"Shape after removing duplicates: {df.shape}")
```

```python
# How many students scored 0 on their final grade?
zero_grades = df[df['G3'] == 0]
print(f"Students with with 0 final grade: {len(zero_grades)}")
print(zero_grades[['G1', 'G2', 'G3', 'absences', 'failures']])
```

```python
absence_amount = df[df['absences'] > 0]
print(absence_amount)
```

```python
df.to_csv('../data/processed/Maths_cleaned.csv', index = False)
print(f"Final cleaned shape: {df.shape}")
print("Saved")
```
