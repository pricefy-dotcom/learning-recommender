## Key Findings
- G1 and G2 are the strongest predictors of final grade.
- Past failures is the most meaningful non-grade predictor.
- Study time has little effect on its own (means more when examined with other information).
- Parental education (Medu [Mom education] and Fedu [Father education]) positively correlate with grades.
- Health, activities, and freetime show no correlation with grades.
- Absences data may have quality issues that could be worth looking into.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the cleaned data
df = pd.read_csv('../data/processed/Maths_cleaned.csv')
print(f"Shape: {df.shape}")
print("Data loaded")
```

```python
fig, axes = plt.subplots(1, 3, figsize = (14, 4))

for i, grade in enumerate(['G1', 'G2', 'G3']):
    axes[i].hist(df[grade], bins = 20, color = 'steelblue', edgecolor = 'black')
    axes[i].set_title(f'{grade} Distribution')
    axes[i].set_xlabel('Grade')
    axes[i].set_ylabel('Number of Students')

plt.suptitle('Grade Distributions Across All Periods', fontsize = 14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (12,8))
numeric_df = df.select_dtypes(include = 'number')
sns.heatmap(numeric_df.corr(), annot = True, fmt = '.2f', cmap = 'coolwarm', center = 0)
plt.title('Correlation Heatmap of All Numeric Features')
plt.tight_layout()
plt.show()
```

```python
# Sort by correlation with G3
correlations = numeric_df.corr()['G3'].sort_values(ascending = False)
print("Factors most correlated with Final Grade (G3):")
print(correlations)
```

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = 'studytime', y = 'G3', data=df, palette = 'Blues')
plt.title('Study Time vs Final Grade')
plt.xlabel('Study Time (1 =< 2 hrs, 2 = 2 - 5 hrs, 3 = 5 - 10 hrs, 4 => 10 hrs')
plt.ylabel('Final Grade')
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = 'failures', y = 'G3', data = df, palette = 'Reds')
plt.title('Past Failures vs Final Grade')
plt.xlabel('Number of Past Failures')
plt.ylabel('Final Grade')
plt.show()
```

```python
plt.figure(figsize = (8, 5))
plt.scatter(df['absences'], df['G3'], alpha = 0.5, color = 'green')
plt.xlabel('Number of Absences')
plt.ylabel('Final Grade')
plt.title('Absences vs Final Grade')
plt.show()
```

```python
plt.figure(figsize = (6, 5))
sns.boxplot(x = 'sex', y = 'G3', data = df, palette = 'Set2')
plt.title('Gender vs Final Grade')
plt.xlabel('Gender')
plt.ylabel('Final Grade')
plt.show()
```
